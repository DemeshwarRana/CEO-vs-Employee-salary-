```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\ragha\Downloads\ceo_data_pay_merged_r3000 (1).csv")
df['ratio_num'] = df['pay_ratio'].str.split(':').str[0].str.replace(',', '').astype(float)
median_overall_ratio = df['ratio_num'].median() 
total_icons = int(median_overall_ratio)
col = 1
rows = (total_icons // col) + 1

x = np.tile(np.arange(col), rows)[:total_icons]
y = np.repeat(np.arange(rows), col)[:total_icons]
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=y, y=x,
    mode='markers',
    marker=dict(symbol='square', size=15, color='#8C97B0'),
    name=f'CEO Pay ({total_icons}x)',
    orientation='h'
))
fig.add_trace(go.Scatter(
    x=[0], y=[0],
    mode='markers',
    marker=dict(symbol='square', size=15, color='#E74C3C'),
    name='Median Worker Pay',
    orientation='h'
))

fig.update_layout(
    title=f"The <span style='font-size:20px;'><b>Pay Gap</b></span> Across {len(df)} US Corporations(2023)<br><sup>Can a corporation be healthy if its base is struggling?",
    xaxis_visible=False,
    yaxis_visible=False,
    plot_bgcolor='white',
    height=500,
    width=1000,
    showlegend=False
)
fig.add_annotation(
    x=0,
    y=0,
    text='Employee Salary',
    showarrow=True,
    arrowwidth=.5,
    arrowcolor='grey',
    ax=0,
    ay=50,
    font=dict(size=12,color='black'),
    align='left',
    bgcolor='white',
    opacity=0.8
)
fig.add_annotation(
    x=90,
    y=0,
    text=f'The Median CEO<br> earns <b>{total_icons}</b> times more<br> than their employee',
    showarrow=True,
    arrowwidth=.5,
    arrowcolor='grey',
    ax=0,
    ay=-60,
    font=dict(color='black',size=12),
    align='right',
    bgcolor='white',
    opacity=0.8)
fig.show()
```
